```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 14500, 9500]
```

```python
plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
# plt.legend()
# plt.tight_layout()
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
plt.pie(
 runs,
 labels=labels,
 colors=colors,
 wedgeprops={'edgecolor': 'black', 'linewidth': 1.9, 'linestyle': '--'}
)
plt.show()
```

```python
explode = [0.1
           , 0, 0, 0] # Only highlight Sachin's slice
plt.pie(runs, labels=labels, explode=explode, colors=colors,shadow=True )
plt.title("Exploded Pie Example")
plt.show()
```
