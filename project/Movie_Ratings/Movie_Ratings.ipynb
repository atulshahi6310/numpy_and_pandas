```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("movies_data.csv")
```

```python
df.head()
```

```python
g = df["Genre"]
g
```

```python
y = df["Year"]
y
```

```python
average_rateing = df.groupby("Genre")["Rating"].mean().round(2)
```

```python
higest_rateing_movies = df.loc[df["Rating"].idxmax()]
```

```python
average_rateing

```

```python
higest_rateing_movies
```

```python
most_popular_year = df["Year"].value_counts().idxmax()
most_popular_year_count = df["Year"].value_counts().max()
```

```python
most_popular_year
```

```python
plt.bar(average_rateing.index, average_rateing.values)
plt.title("Average Rating by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
g.value_counts().plot(kind = 'pie' , autopct="%1.1f%%")
```

```python
df.groupby("Year")["Rating"].mean().plot(kind="line", marker="o")
plt.title("Ratings Trend by Year")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.grid(True)
plt.show()
```
