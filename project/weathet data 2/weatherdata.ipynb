### **Problem Statement (Day 8 - Weather Data Analysis)**

You are provided with daily **weather data** of a city.
Your task is to:

1. Analyze **temperature patterns** (min, max, average).
2. Find the **day with the highest temperature**.
3. Plot a **line graph of temperature vs. date**.
4. Create a **bar plot of rainfall over the days**.
5. Show a **correlation heatmap** between temperature, humidity, and rainfall.




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("weather_day8.csv")
```

```python
df.head()
```

```python
avg_temp = df["Temperature (¬∞C)"].mean()
max_temp = df["Temperature (¬∞C)"].max()
min_temp = df["Temperature (¬∞C)"].min()

print("\nAverage Temp:", avg_temp)
print("Max Temp:", max_temp)
print("Min Temp:", min_temp)
```

```python
hottest_day = df.loc[df["Temperature (¬∞C)"].idxmax()]
```

```python
hottest_day
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Date"], df["Temperature (¬∞C)"], marker='o', color="red", label="Temperature")
plt.xticks(rotation=45)
plt.title("Daily Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (¬∞C)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Date", y="Rainfall (mm)", data=df, color="blue")
plt.xticks(rotation=45)
plt.title("Daily Rainfall")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[["Temperature (¬∞C)", "Humidity (%)", "Rainfall (mm)"]].corr(),
            annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
