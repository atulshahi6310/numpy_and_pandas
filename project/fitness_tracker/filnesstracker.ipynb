```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("fitness_data.csv")
```

```python
df
```

```python
total_steps = df["Steps"].sum()
```

```python
total_steps
```

```python
total_calories =df["Calories"].sum()
```

```python
total_calories
```

```python
avg_sleep = df["Sleep Hours"].mean()
```

```python
avg_sleep 
```

```python
most_step = df.loc[df["Steps"].idxmax(),"Day"]
```

```python
most_step
```

```python
max_calories_day = df.loc[df["Calories"].idxmax(), "Day"]
```

```python
max_calories_day
```

```python
correlation = np.corrcoef(df["Sleep Hours"], df["Steps"])[0, 1]
```

```python
correlation
```

```python
plt.plot(df["Steps"], df["Calories"]  )
plt.tight_layout()
plt.grid()
plt.title("Steps & Calories Trend Over Days")
plt.xlabel("Day")
plt.ylabel("Count")

plt.show()
```

```python
plt.plot(df["Steps"], df["Sleep Hours"]  )
plt.tight_layout()
plt.grid()
plt.title("Steps & sleep Trend Over Days")
plt.xlabel("Day")
plt.ylabel("Count")

plt.show()
```

```python

plt.bar(df["Day"], df["Sleep Hours"], label="Sleep per day")
plt.title("Sleep Hours Comparison")
plt.xlabel("Day")
plt.ylabel("Sleep Hours")
plt.legend()
plt.show()
```

```python
sns.scatterplot(x = df["Sleep Hours"] , y = df["Steps"] , data = df , s =100 , ) 
plt.grid()
```
